We downloaded the MNIST database which is split into 3 parts: a training, test, and validation part. The samples in each subset contain an image (an array of 784 pixel intensity - 28x28 image), and a label (a numerical description of the image). The labels represent "one-hot" vectors, i.e. a vector with all zeros, except for a 1 in the corresponding position for the given label.

```python
# The images are stored in the following Tensor:
images = mnist.train.images

# The labels of each image are stored in:
labels = mnist.train.labels
```

```python
import tensorflow as tf

# A placeholder for any number of images (given with None), each with 784 pixels
x = tf.placeholder(tf.float32, [None, 784])

W = tf.Variable(tf.zeros([784, 10]))
b = tf.Variable(tf.zeros([10]))

y = tf.nn.softmax(tf.matmul(x, W) + b)
```

Let's use a cross-entropy cost function that needs to be minimized.

```python
# A placeholder for the correct labels (any number of samples, each with 10 digits for the 'one-hot' vector)
y_ = tf.placeholder(tf.float32, [None, 10])

# The cross-entropy
# reduce_sum - sum the elements in the second dimension (because of reduction_indices=[1])
# reduce_mean - compute a mean over all samples in the batch
cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y), reduction_indices=[1]))

# Next, tf will use the backprop algorithm to determine how the variables affect the cost function.
# We an use a gradient descent optimizer to modify the variables and reduce the cost
# tf adds the backprop and the gradient to the graph behind the scene
train_step = tf.train.GradientDescentOptimizer(0.5).minimize(cross_entropy)
```

```python
# Run the model in an InteractiveSession
sess = tf.InteractiveSession()

# Initialize variables
tf.global_variables_initializer().run()

# Run the train_step for 100 times
for _ in range(100):
    batch_xs, batch_ys = mnist.train.next_batch(100)
    sess.run(train_step, feed_dict={x: batch_xs, y: batch_ys})
```

In each training step we use a batch of random 100 data points from the training set. This is called _stochastic training_, in this case _stochastic gradient descent_. Ideally, we would like to use all our data in every training step, but that's expensive. Using a smaller subset is cheaper, but has similar benefits. 

```python
# Evaluate the model
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))

# Convert the output list of booleans to a list of float 1-s and 0-s
# Is this accuracy on the training set? or the batch? or what?
accuracy = tf.reduce_mean(tf.cast(correct_prediction), tf.float32)

# Print the accuracy on the test set
print(sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels}))
```
